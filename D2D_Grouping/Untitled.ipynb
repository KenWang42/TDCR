```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
%matplotlib inline
```

```python
nMTCD = 1000
MTCD_Position = pd.read_csv(f'../MTCD_Position/MTCD_position_{nMTCD}.csv')
```

```python
X = np.vstack((MTCD_Position.x, MTCD_Position.y)).T
X.shape
```

```python
Z = sch.linkage(X, 'complete')
```

```python
clusters = sch.fcluster(Z, 100, criterion='distance')
```

```python
np.unique(clusters)
```

```python
plt.figure(figsize=(20, 20))
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='prism')
plt.show()
```

```python
df = MTCD_Position.assign(clusters=list(clusters))
df.head()
```
